```python
df = pd.read_csv("Terrorism_copy_globe.csv",delimiter=";")
df.sample()
```

```python
#A choice had to be made: Czechoslovakia split up in parts. As Tsjechie is larger than Slovakia, I will set it to CZE
#Another choice: Yugoslavia. I opted to assign it to Serbia. If someone does not agree let me know
Problematic_countries = {u"West Germany (FRG)":u"DEU", u"East Germany (GDR)":u"DEU", u"Venezuela":"VEN",u"Democratic Republic of the Congo":u"COD", u"West Bank and Gaza Strip":u"PSE", u"Russia":u"RUS",u"Syria":u"SYR", u"Iran":u"IRN", u"Bolivia":u"BOL", u"South Yemen":u"YEM", u"North Yemen":u"YEM", u"Taiwan":u"TWN",u"Brunei":u"BRN", u"Czechoslovakia":u"CZE", u"Zaire":u"COD", u"South Vietnam":u"VNM",u"People's Republic of the Congo":u"COG", u"Yugoslavia":u"SVK",u"Rhodesia":u"ZWE",u"South Korea":u"KOR",u"Tanzania":u"TZA", u"Vatican City":u"VAT",u"Soviet Union":u"RUS",u"New Hebrides":u"VUT", u"Swaziland":u"SWZ", u"Laos":u"LAO",u"Republic of the Congo":u"COG",u"Falkland Islands":u"FLK",u"Ivory Coast":u"CIV", u"Moldova":u"MDA", u"Bosnia-Herzegovina":u"BIH",u"Vietnam":u"VNM", u"Slovak Republic":u"SVK", u"Macedonia":u"MKD", u"Czech Republic":u"CZE", u"North Korea":u"PRK", u"Macau": u"MAC", u"Kosovo":u"XKX",u"St. Kitts and Nevis":u"KNA", u"East Timor":u"TLS",u"St. Lucia":u"LCA",u"Serbia-Montenegro":u"SRB", u"International":u"INTRN"}

def CCGrabber(x):
    #First, we deal with the problematic ones
    if x in Problematic_countries:
        return Problematic_countries[x]
    else:
        try:
            return pc.countries.get(name=x).alpha_3
        except:
            print(x) #Consider the names printed here, and look at the corresponding name in the word.json file
            return x[3]

df["CountryCode"] = df.country_txt.apply(lambda x: CCGrabber(x),1)
```

```python
df.to_csv('Globe_data_Country_code', encoding='utf-8', index=False)
```

```python
#We now procceed by making data sets with the desired properties
#We will always split by year (for the slider) and by country (for obvious reasons)

amount_of_deaths = df.groupby(["CountryCode", "iyear"])["nkill"].sum()
amount_of_attacks = df.groupby(["CountryCode","iyear"])["eventid"].count()
## Change from a groupby object into a dataframe we can work with
amount_of_deaths =amount_of_deaths.reset_index()
amount_of_attacks = amount_of_attacks.reset_index()
```

```python
#Now, we add them together!
df_for_globe = pd.merge(amount_of_deaths,amount_of_attacks,on=["CountryCode","iyear"])
df_for_globe = df_for_globe.rename({"eventid":"nattacks"},axis="columns")
```

```python
df_for_globe.to_csv('GlobeData.csv', encoding='utf-8', index=False)
```
