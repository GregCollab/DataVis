```python
import os
import numpy as np
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
import csv
from wordcloud import WordCloud,STOPWORDS
```

```python
text = pd.read_excel("WordCloudData.xlsx")
```

```python
stopwords = set(STOPWORDS)

textTest = text["targtype1_txt"].tolist()
textTest = " ".join(textTest).lower()
```

```python
wc=WordCloud(background_color="rgba(255, 255, 255, 0)", mode="RGBA", colormap='Set2',collocations=False)
wc.generate(textTest)
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wc,interpolation='bilinear')
plt.tight_layout(pad = 0)
plt.axis("off")
```

```python
wc.to_file("Wordcloud.png")
```
